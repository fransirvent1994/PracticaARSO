## Ping de máquina host y guest

El ping es la conexión (envío de paquetes, por ejemplo) entre tu PC y la VM.

<h3><b><u> IP de tu máquina host

Normalmente conseguimos la IP mediante el comando en terminal:

- <b> ifconfig </b> (Linux)
o bien
- <b> ipconfig </b> (powershell/CMD)

Tras introducir el comando anterior, debería salir una serie de valores que muestren la IP o IPs de cada conexión.

Usando el comando en CMD o powershell:

<b>ping <IP de máquina virtual>

Usando el comando:

<b>ping -c1 "Ip de PC"

En mi caso estoy usando una conexión de datos con mi móvil, por lo que la IP variará si hacéis lo mismo y luego cambiais a otra conexión de internet.

<b>De esta manera tenemos ya establecida la conexión entre máquina host y máquina virtual.

<b><u>Pequeño tip:</b></u>

Si en la máquina virtual no sabes cual es tu IP entre todas las redes que salen, prueba con el comando:

- ip route
