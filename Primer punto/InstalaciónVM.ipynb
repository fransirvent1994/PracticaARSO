### Instalación Máquina virtual

Para instalar la máquina virtual tenemos 2 opciones:

- VM de Oracle Virtual Box
- Partición de disco, activación de hipervirtualización y crear la máquina.

Personalmente, mucho más sencillo la primera opción.

<u>Link de instalación oficial: <a href= "https://www.virtualbox.org/">Instalar Virtual Box</a>

La instalación no tiene mucho misterio. El botón download se encuentra en la misma web, es gratuito, y la configuración es a gusto de cada uno.

Para continuar, habrá que descargar el iso de Debian. Si hacéis click en el enlace de los recursos, no funcionará porque el enlace está partido. 

<u>Aquí lo dejo junto y arreglado.
<a href="https://cdimage.debian.org/debian-cd/current/amd64/iso-cd/debian-12.10.0-amd64-netinst.iso">Descarga Debian 12.10</a>

El enunciado pide al menos 15 gigas para la máquina virtual. Si vas sobrado, pilla 20 y 2 núcleos. Explico como funciona la configuración. Es posible que en otros sistemas se vea cosas de distinta forma, pero al uso será lo mismo todo.

La imagen siguiente muestra lo que tengo en mi máquina virtual. Tengo 2 creadas. A la derecha aparecen iconos que son para la configuración. Si no tenías este programa, debería salir vacío.

<h3><b><u>Crear una máquina virtual

Irás al botón "Nueva" y configuras la máquina de este modo:



En mi caso, ya tengo la configuración hecha antes en otra máquina, así que me saldrá error.

En el enunciado pide que el nombre de la máquina sea ARSO20242, así como está escrito en la configuración.

La instalación de la iso va en el apartado "imagen ISO". Automáticamente se convertirá el resto en el Sistema Operativo de Linux.

La siguiente imagen se encuentra en el apartado "instalación desatendida", y es para abordar el punto de nombre de usuario y contraseña.

En el enunciado se pide el nombre de usuario de la cuenta de la UOC y habrá que crear una contraseña. Así que, donde pone "nombre de usuario", deberás poner tu nombre de usuario de la UOC.

En el apartado "hardware" se encuentra la memoria y el número de procesadores. Aquí sugeriría dejar la memoria base como está y usar 2 procesadores, para mayor fluidez en la máquina.

En la pestaña de "disco duro" se encuentra la capacidad de memoria, que en el enunciado dice 15 gigas, y con eso debería sobrar, pero yo puse 20 por ir sobrado por si acaso.

Una vez configurado todo, si está todo bien, debe estar activado el botón "terminar". Al clickar se creará la máquina virtual con el nombre de usuario y la contraseña ya hechos.

De esta manera, una vez accedes a la máquina, la configuración será automática para esto. El resto de configuraciones tardarán un rato. Para acceder a la máquina y proceder con las instalaciones, pulsa el botón iniciar del programa de virtual box.

Debian se debería ver así una vez ya configurado todo. Si tenéis aplicaciones de escritorio no creo que sea problema, pero lo que se pide en el enunciado es todo desde terminal.

Pulsando el botón "windows" accedemos a aplicaciones, entre ellas está la terminal.

Normalmente el usuario creado debería tener acceso a sudo para acceder al root, pero, si no es así, aquí dejo el comando necesario para crear privilegios a un usuario.

<b>sermod -aG sudo "tu nombre de usuario"
