<h1>Git

Si no tienes instalado git, te explico como instalarlo y trabajar con él.

Para empezar, te irás a la terminal de debian, y aplicarás el comando:

sudo apt update
sudo apt install git -y

Si está bien instalado, usaremos el comando:

- <b>git --version

Así comprobaremos qué versión está instalada (por defecto, la más reciente)

Una vez hecho esto, usaremos:

- <b> git clone https://github.com/jestebangr/prac20242-orig.git

Que viene a ser el repositorio en el que se encuentran los materiales para esta práctica.

Con este comando se clonará el repositorio en el directorio en el que estamos, así que, cuando esté instalado, accederemos a él mediante:

- cd "prac20242-orig.git"

Puedes usar el comando "ls" para comprobar si se encuentra entre los archivos.

Una vez dentro del directorio, podremos usar el comando "ls" para ver qué contiene, y el comando "cat" para acceder a la lectura de los documentos.

<h3> Servicio Postgres

Mediante el comando:

- cat docker-compose.yml

Podremos ver qué contiene mediante lectura, y veremos que es la configuración de base de datos.

Mediante el comando:

- nano docker-compose.yml

Accedemos para editar el documento, añadiendo lo necesario según el enunciado.

En user y Password debes tener tu nombre de usuario y contraseña creados en la máquina virtual previamente.

Una vez hecho esto, usamos comando:

- <b> docker compose up --build

Esto ejecuta el contenedor docker del archivo, así que, mientras está ejecutandose, abriremos otra terminal y usaremos el comando:

- docker ps
